# Case: App de delivery que guarda os dados dos clientes em uma planilha e precisamos manipulá-los para melhorar a análise em estudos futuros.

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/banco_clientes.csv')
df
```

Padronizar os nomes em letra maiúscula.



```python
df['nome'] = df['nome'].str.upper()
df
```

Padronizar a coluna de telefones para o padrão: XX XXXXX-XXXX

```python
df.telefone.str.replace('(', '', regex = False)
```

```python
df.telefone.str.replace('(', '', regex = False).str.replace(')', '', regex = False)
```

```python
df['telefone'] = df.telefone.str.replace('(', '', regex = False).str.replace(')', '', regex = False)
```

Criar uma coluna apenas com os CEPs

```python
df.endereco.str.extract('(\d{5}-\d{3})')
```

```python
df['cep'] = df.endereco.str.extract('(\d{5}-\d{3})')
df
```

```python
df.endereco.str.replace('\d{5}-\d{3}', '', regex = True)
```

```python
df['endereco'] = df.endereco.str.replace('\d{5}-\d{3}', '', regex = True).str.strip()
df
```

Ter uma coluna para rua e outra para o número do endereço do cliente.

```python
df.endereco.str.split('\s(?=\d+$)', expand = True)
```

```python
df.endereco.str.replace(',', '').str.split('\s(?=\d+$)', expand = True)
```

```python
df[['rua', 'numero']] = df.endereco.str.replace(',', '').str.split('\s(?=\d+$)', regex = True, expand = True)
df
```

```python
df = df.drop(['endereco'], axis = 1)
df
```

# Vendas Online
Este notebook contém a análise de resultados de um evento com clientes. Foram coletados dados de clientes que mais gastaram durante os 5 dias do evento. O cliente com maior compra foi identificado, recebeu um prêmio e  pode ajudar a empresa a criar novas estratégias de vendas.

```python
import pandas as pd
```

```python
dados = pd.read_json('/content/dados_vendas_clientes.json')
dados
```

```python
dados = pd.json_normalize(dados['dados_vendas'])
dados
```

```python
colunas = list(dados.columns)
colunas
```

```python
dados = dados.explode(colunas[1:])
dados
```

```python
dados.reset_index(inplace = True, drop = True)
dados
```

```python
dados.info()
```

```python
col_numerica = ['Valor da compra']
```

```python
import numpy as np
```

```python
dados['Valor da compra'] = dados['Valor da compra'].apply(lambda x: x.replace('R$ ', '').replace(',' , '.').strip())
dados['Valor da compra'] = dados['Valor da compra'].astype(np.float64)
```

```python
dados.head()
```

```python
dados['Cliente'] = dados['Cliente'].str.lower()
dados['Cliente']
```

```python
dados['Cliente'] = dados['Cliente'].str.replace('[@\0-9\_\']', ' ', regex = True)
```

```python
dados['Cliente']
```

```python
import datetime
```

```python
dados['Data de venda'] = pd.to_datetime(dados['Data de venda'], format = '%d/%m/%y')
```

```python
total_compras = dados.groupby(['Cliente'])['Valor da compra'].sum()
total_compras
```
